```python
# Importeren van bibliotheken
import math
import numpy as np
import matplotlib.pyplot as plt
import constanten as C

plt.style.use("classic")
```

```python
# Doorbuiging zwaarste punt in 3D
I = 2*C.H_WEGDEK**3/12
E = C.E_MODULUS
Q = C.VERDEELDE_BELASTING/2+C.EIG_GEWICHT_BRUG
P = C.PUNT_BELASTING/2

D = lambda L: (5*Q*L**4)/(384*E*I) + (5*P*L**3)/(48*E*I)

x = np.arange(0, 0.4, 0.0001)
plt.plot(x, [D(i)*1000 for i in x])
plt.title("Doorbuiging in 3D meest belaste punt")
plt.ylabel("Doorbuiging [mm]")
plt.xlabel("Ophangpunten tov midden van brug [m]")
```
