```python
# Importeren van bibliotheken
import math
import numpy as np
import matplotlib.pyplot as plt

from berekeningen_ramses import Brug

plt.style.use("classic")
```

```python
import genetic_algo as ga
```

```python
# Creeer van populatie
population = [ga.genRandomBrugB() for i in range(100)] # Populatie van 100
```

```python
# Random algo GO
def evolvePop(population):
    for i in range(100): # 1000 generations
        print('.', end="")
        population = ga.kill_population(population)
        population = ga.regen_population(population,
                                        ga.genRandomBrugB,
                                        ga.genFromParentsB)
    return population

%time population = evolvePop(population)
```

```python
best = ga.find_best(population)
ga.print_brug(best[1])
print(f"Score brug: {best[0]}")
```
